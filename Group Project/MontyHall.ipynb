```python
#from scipy.stats import qmc
import matplotlib.pyplot as plt
import random
```

```python
#lists to store probability values after each iteration

switchWinProbability = []
stickWinProbabilitiy = []

doors = [True, False, False]

#Monte Carlo Simulation
def MontyHall(NSimulations):                 #input number of simulations
    switchWinCounter = 0
    stickWinCounter = 0

    for i in range(NSimulations):              #iterate through N # of Simulations
        
        random.shuffle(doors)                   #randomize indices         

                                                #initial choice from 3 doors
        k = random.randrange(0,2)               #pick 0,1 or 2

        if((doors[k]) != True):                 #if kth index in doors is not the prize
            switchWinCounter +=1                #should've switched
        else:                                   #if kth index in doors is the prize
            stickWinCounter +=1                 #should've stuck
        
                                                        #calculate probability every iteration
        winBySwitch = switchWinCounter/(i + 1)          #add 1 because i starts with 0
        winByStick = stickWinCounter/(i + 1)
        

        #update list values
        switchWinProbability.append(winBySwitch)
        stickWinProbabilitiy.append(winByStick)
    
    return switchWinProbability[-1], stickWinProbabilitiy[-1]


#need two variables to assign to, because 2 return values
switch, stick = MontyHall(1000)

print('switch win probability:', switch)
print('stick win probability', stick)

plt.plot(switchWinProbability)
plt.plot(stickWinProbabilitiy)
```

```python
def insertion_sort(list_1):

    for i in range(1, len(list_1)):
        
        temp = list_1[i]

        j = i -1                                #define j

        while j >= 0 and temp < list_1[j]:

            list_1[j + 1] = list_1[j]

            j -= 1                                  #go down an index     

        list_1[j + 1] = temp

    return list_1
        
```

```python
my_list = [6,3,-1,5,7, 10]
insertion_sort(my_list)
```

```python
def climb_stairs(n):
    if n < 2:
        return 1                #only one way to climb stair

    else:
        return climb_stairs(n - 1) + climb_stairs(n - 2)
```

```python
print(climb_stairs(3))
print(climb_stairs(5))
```

```python
import matplotlib.pyplot as plt
import numpy as np


def polygon(n):

    theta = 2 * np.pi/n
    radius = 1

    xs = []
    ys= []

    for i in range(n + 1):
        xs.append(radius*np.cos(i*theta))
        ys.append(radius*np.sin(i*theta))

    plt.figure(figsize=(5,5))
    
    for i in range(n-1):
        for j in range(i+1, n):
            plt.plot([xs[i], xs[j]], [ys[i], ys[j]], color = 'k')
```

```python
polygon(10)
```

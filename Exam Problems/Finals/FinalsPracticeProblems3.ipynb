```python
import matplotlib.pyplot as plt

class Square():
    
    def __init__(self, C, l):
        if (isinstance(C, tuple) and len(C) == 2):
            pass
        else: 
            raise(TypeError('C has to be a list of tuples and l has to be a float'))

        self.__C = C  
        self.__V = [(C[0] - l/2, C[1] - l/2), (C[0] + l/2, C[1] -l/2), (C[0] + l/2, C[1] + l/2), (C[0]-l/2, C[1] + l/2)]

    def vertex(self, n):
        return self.__V[n-1]

    def sideLength(self):
        return abs(self.__V[0][0] - self.__V[1][0])                        #only the x components 
           
    def show(self, color = 'k'):
        plt.fill([x for x, y in self.__V], [y for x, y in self.__V], color)                                                         #two lists using list comprehension

    def __str__(self):
        return str(self.__V)
```

```python
##Test Case

a = Square((0,0), 1)
print(a.vertex(3))
print(a.sideLength())
print(a)
a.show()
```

```python
def bubble_sort(list_1):
    is_Sorted = False
    while not is_Sorted:
        is_Sorted = True
        for i in range(len(list_1) -1):
            if list_1[i] < list_1[i + 1]:
                list_1[i], list_1[i + 1] = list_1[i + 1], list_1[i]
                is_Sorted = False
    return list_1
```

```python
#Test Case
my_list = [6,3,-1,5,7,10]

print(bubble_sort(my_list))
```

```python
import matplotlib.pyplot as plt
import numpy as np

def draw(n):
    points = np.linspace(0, 1, n)                   #make a list

    plt.figure(figsize=(5,5))                       #plot size; don't really need    

    for i in range(n):
        plt.plot([0,points[i]], [points[n- 1 - i], 0], color = 'k')
```

```python
draw(20)
```

```python
#Question 4
#Write the function getNumbers(str_1) which receives a string and returns the list of numbers inside that string.

def getNumbers(inputString):
    outputList = []
    temp = ''

    for i in range(len(inputString)):
        if inputString[i].isdigit() or inputString[i] == '+' or inputString[i] == '-'\
             or inputString[i] == '.':

            temp += inputString[i]
        
        #i == len(inputString) - 1) means the last index
        if(inputString[i] == ' ' or i == len(inputString) - 1) and len(temp):
            outputList.append(float(temp))
            #make temp empty again
            temp = ''


    return outputList
```

```python
#Question 4 Solution

a = getNumbers(' a=0.    b=-0.08642     c=0.3523')
print(a)
```

```python
#Question 5

# Read the file in Python and plot the data against time in two subplots as shown below.
#First thirteen lines are information and then there are three columns, 
# first one is time, second and third columns are displacements in the x- and y-directions. 

import matplotlib.pyplot as plt

f = open('./data.txt', 'r')
content = f.readlines()                     #readlines() gives a list of strings, line by line
f.close()
skipLines = 13

import numpy as np
data = []
for i in content[skipLines - 1:]:
    data.append(getNumbers(i))              #taking the numbers out of the strings
                                            #e.g. data = getNumbers(data[15]) = [[0, -0.8, 0.3]]
data = np.array(data)                       #e.g. data = [...[0, -0.8, 0.3], [0.1, -0.1, 0.3], ...]

print(data)

plt.figure(figsize=(14,5))
plt.subplot(1,2,1)
plt.plot(data[:,0], data[:,1])              #time by x
plt.xlabel('Time (s)')
plt.xlim([0,60])
plt.ylabel('Displacement x-direction (mm)')

plt.subplot(1,2,2)
plt.plot(data[:,0], data[:,2])              #time by y
plt.xlabel('Time (s)')
plt.xlim([0,60])
plt.ylabel('Displacement x-direction (mm)')

plt.tight_layout()
plt.show()
```

```python
#Question 6

#A. Function that swaps elements

#Method 1
def swap(inputList, x, y):
    inputList[x], inputList[y] = inputList[y], inputList[x]
    return inputList

#Method 2
def swap2(inputList, x, y):
    temp = inputList[x]
    inputList[x] = inputList[y]
    inputList[y] = temp
    return inputList
    
```

```python
#Test Case
myList = [2,0,4,1]
print(swap(myList, 0,3))
print(swap2(myList, 0,3))
```

```python
#B. Selection sort
#selection sort selects a pivot, and goes through elements to see if its greater
#if not, then new number is the 'greatest'
def SelectionSort(inputList):
    
    for j in range(len(inputList)):
        maxIndex = j
        for i in range(j + 1, len(inputList)):              #j + 1 to check with the next element
            if(inputList[maxIndex] < inputList[j]):
                maxIndex = i
        inputList[j], inputList[maxIndex] = inputList[maxIndex], inputList[j]

    return inputList
        
#2nd method, you can have 2 lists: sorted and unsorted list
#the sorted list is empty, you compare the max number with all the elements in the unsorted list
# , and take it out of the unsorted list and put it into the sorted list

```

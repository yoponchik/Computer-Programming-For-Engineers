```python
#1
%matplotlib inline

def first_derivative(f,a,b):
    x = np.arange(a,b)
    plt.plot(f,x)
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.show()

f = lambda x: np.sin(x)
first_derivative(f,0, 2*np.pi)
```

```python
#2
def is_prime(n):
    if(n < 2):                             #base case
        raise Exception("Error: Need input greater than or equal to 2.")   
    else:
        i = 2
        while(n/2 >= i):                    #start dividing from 2 and greater
            if(n%i == 0):                   #if can't be divisible by itself
                return False                  #not prime    
            i += 1
        return True

#Test Case
print(is_prime(10))
print(is_prime(2))
print(is_prime(997))

#3 
def prime_number(n):
    for i in range(n):
        if(i < 2):
            continue
        if(is_prime(i)):
            print(i)

prime_number(1000)
```

```python
from math import pi
import numpy as np
import matplotlib.pyplot as plt

r = 10 # Radius of the wheel
phi = np.arange(0, 4*pi, pi/180) # Rotation from 0 to 4π with the step size of 1°

# Coordinates of the cycloid
x = r * (phi - np.sin(phi))
y = r * (1 - np.cos(phi))
plt.figure(figsize=(4, 3))
plt.plot(x, y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

```python
x = np.linspace(-4, 4, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x,y)
Z = X * Y * (X**2 - Y**2) / (X**2 + Y**2)

from mpl_toolkits import mplot3d
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(1, 2, 1, projection='3d')
ax.plot_surface(X, Y, Z)
ax.set_title('Surface plot')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax = fig.add_subplot(1, 2, 2, projection='3d')
ax.plot_wireframe(X, Y, Z)
ax.set_title('Wireframe plot')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.tight_layout()
plt.show()
```

```python
def my_polygon(n):
    from math import pi
    import numpy as np
    import matplotlib.pyplot as plt
    
    r = 1 # Radius of the polygon
    theta = np.arange(0, 2*pi+2*pi/n, 2*pi/n) # Angular coordinate of the vertices

    # Coordinates of the vertices
    x = r * np.sin(theta)
    y = r * np.cos(theta)
    plt.figure(figsize=(4, 4))
    plt.plot(x, y)
    plt.title(f'My Regular Polygone with {n} sides')
    plt.xlabel('x axis')
    plt.ylabel('y axis')
    plt.axis('equal')
    plt.show()
    return

my_polygon(5)
my_polygon(10)
```

```python
def my_fun_plotter(f, x):
    import matplotlib.pyplot as plt
    y=[f(i) for i in x] # List comprehension
    plt.plot(x, y)
    plt.xlabel('x axis')
    plt.ylabel('y axis')
    plt.axis('equal')
    plt.show()
    return

my_fun_plotter(lambda x : np.sqrt(x) + np.exp(np.sin(x)), np.linspace(0, 2*np.pi, 100))
```

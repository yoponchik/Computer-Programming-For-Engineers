```python
def getNumbers(str_1):
    numbers = []
    tmp = ''

    for i in range(len(str_1)):
        # Finding digit characters, signs, and points
        if str_1[i].isdigit() or str_1[i]=='+' or str_1[i]=='-'\
        or str_1[i]=='.':
            tmp += str_1[i]
        
        # Joining the obtained characters, if available, at each space
        # or at the end of the string
        if (str_1[i]==' ' or i==len(str_1)-1) and len(tmp):
            numbers.append(float(tmp))
            tmp=''
            
    return numbers

# Test case ---------------------------------------
a=getNumbers('  a=0.    b=-0.08642     c=0.35256')
a
```

```python
f=open('data.txt','r')
contents = f.readlines()
f.close()
skipLines = 13

#-------------------------------------------
import numpy as np
data = []
for i in contents[skipLines-1:]:
    data.append(getNumbers(i))
data = np.array(data)

#-------------------------------------------
import matplotlib.pyplot as plt
plt.figure(figsize=(14,5))
plt.subplot(1,2,1)
plt.plot(data[:,0],data[:,1])
plt.xlabel('Time (s)')
plt.xlim([0,60])
plt.ylabel('Dipslacement x-direction (mm)')

plt.subplot(1,2,2)
plt.plot(data[:,0],data[:,2])
plt.xlabel('Time (s)')
plt.xlim([0,60])
plt.ylabel('Dipslacement y-direction (mm)')

plt.tight_layout()
plt.show()
```

```python
def swap(list_1, ele_1, ele_2):
    list_1[ele_1], list_1[ele_2] = list_1[ele_2], list_1[ele_1]
    return list_1

# Test case -----------------------------------------------------
my_list = [2,0,4,1]
print(swap(my_list,0,3))
```

```python
def selection_sort(list_1):
    for i in range(len(list_1)):
        max_index = i
        for j in range(i+1,len(list_1)):
            if list_1[max_index] < list_1[j]:
                max_index=j
                
        list_1[i], list_1[max_index] = list_1[max_index], list_1[i]
        print('Step ',i+1,' = ', list_1)
    return list_1

my_list = [6, 3, -1, 5, 7, 10]
selection_sort(my_list)
```
